```python
from IPython.display import display_html
def display_side_by_side(*args):
    html_str=''
    for df in args:
        html_str+=df.to_html()
    display_html(html_str.replace('table','table style="display:inline"'),raw=True)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
prv=pd.read_csv('lgg_tcga_clinical_data.tsv', sep='\t')
atls=pd.read_csv('lgg_tcga_pan_can_atlas_2018_clinical_data.tsv', sep='\t')
```

<br/><br/>
## <u>Comparison of ' Mutation Count vs Fraction of Genome Altered '</u>

```python
%matplotlib Inline
%config InlineBackend.figure_format = 'svg'
f, (ax1, ax2) = plt.subplots(1,2,figsize=(10, 4.8))

f.suptitle ('Mutation Count vs Fraction of Genome Altered')

yatls = atls['Mutation Count']
xatls = atls['Fraction Genome Altered']
ax1.set_xlabel('Fraction Genome Altered')
ax1.set_ylabel('# of Mutation')
ax1.scatter(xatls,yatls)

yprv = prv['Mutation Count']
xprv = prv['Fraction Genome Altered']
ax2.set_xlabel('Fraction Genome Altered')
ax2.set_ylabel('# of Mutation')
ax2.scatter(xprv,yprv)

plt.show()
```

## <u>Comparison of Deletion and Amplification Genes</u><br/>


```python
mutateAtls=pd.read_csv('Mutated_Genes_Atlas.txt', sep='\t')
mutatePrv = pd.read_csv('Mutated_Genes_Provisional.txt', sep='\t')
```

```python
mutateAtls1 = mutateAtls.dropna(subset=['# Mut']).where(mutateAtls['# Mut'] > 120).dropna(how='all')[['Gene','# Mut','#','Freq','Is Cancer Gene (source: OncoKB)']]
```

```python
mutatePrv1 = mutatePrv.dropna(subset=['# Mut']).where(mutatePrv['# Mut'] > 60).dropna(how='all')[['Gene','# Mut','#','Freq','Is Cancer Gene (source: OncoKB)']]
```

```python
print('Mutated Genes Atlas                                     Mutated Genes Provisional')
display_side_by_side(mutateAtls1,mutatePrv1)
```

We can see that we got <b>4 genes</b> above are the <b>same for both cases of Brain Cancer</b> <br/>
They are present nearly <b>over 20%</b> and expand to <b>more than 70%</b> of the cases<br/>
Also,they are <b>all marked as Cancerous Genes</b>

<br/><br/>
## <u>Comparison of mutated genes in the RNA</u>

```python
cnatls=pd.read_csv('CNA_Genes_Atlas.txt', sep='\t')
cnaprv = pd.read_csv('CNA_Genes_Provisional.txt', sep='\t')
```

```python
cnatls1 = pd.DataFrame(cnatls.groupby(cnatls['CNA'])['#'].sum()).rename_axis('CNA Atlas',axis = 1)
```

```python
cnaprv1 = pd.DataFrame(cnaprv.groupby(cnaprv['CNA'])['#'].sum()).rename_axis('CNA Provisional',axis = 1)
```

```python
display_side_by_side(cnatls1,cnaprv1)
```

Above, the comparison of <b>mutated genes in the RNA</b> <br/><br/>
We can see that the <b>Provisional Dataset</b> got much <b>more Amplification and Deletion</b> for the analysed genes<br/><br/>
As we can see for <b>Amplifications</b> the <b><i>Atlas Dataset</i></b> got <b>55822</b> against <b>78670</b> for the <b><i>Provisional Dataset</i></b><br/>
About the <b>Deletions</b> the <b><i>Atlas Dataset</i></b> got <b>37498</b> against <b>66232</b> for the <b><i>Provisional Dataset</i></b>

## <u>Analysis of Tumor Sites for ' Brain Lower Grade Glioma (TCGA, Provisional) '</u>
### <u><i>No data available for ' Brain Lower Grade Glioma (TCGA, PanCancer Atlas) '</i></u>

```python
%config InlineBackend.figure_format = 'svg'


Site = prv['Tumor Site']
sums = prv.groupby(Site)['Number of Samples Per Patient'].sum()

explode = (0.7, 0.5, 0.3, 0.1, 0.1, 0.1, 0.1)

plt.pie(sums.sort_values(), labels = sums.index, startangle=90, autopct='%1.1f%%',
        labeldistance = None, explode=explode, pctdistance=1.2, textprops={'size': 'smaller'}, shadow = True, counterclock=False)
plt.title('Tumor Sites', pad = 80)
plt.legend(loc="center left",bbox_to_anchor=(1.5, 0, 0, 1))
plt.show()
```

The pie above show us the different sites where <b>Brain Lower Grade Glioma</b> where found for the <b>Provisional dataset</b> <br/><br/>
We can see in pink that <b>over 50%</b> of the cancer mentioned above are located in the <b>Supratentorial, Temporal Lobe</b> <br/>
The brown section that <b>nearly 30%</b> are located in the <b>Supratentorial, Parietal Lobe</b><br/>
The <b>Supratentorial Occipital Lobe</b> represent <b>9%</b> of the case of Brain lower Grade Glioma<br/><br/>
Finally, the <b>four other tumor sites</b> represent  <b>between 0.2 and 1.4 percent </b>each, located in:
<ul><b>
    <li>Posterior Fossa, Brain Stem</li>
    <li>Posterior Fossa, Cerebellum</li>
    <li>Supratentorial, Frontal Lobe</li>
    <li>Supratentorial, Not Otherwise Specified</li>
